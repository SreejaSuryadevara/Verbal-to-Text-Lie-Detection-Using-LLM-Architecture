```python
import matplotlib.pyplot as plt

# Function to read accuracies from the text file
def read_accuracies(file_path):
    accuracies = {}
    with open(file_path, 'r') as file:
        for line in file:
            # Split each line by colon ':'
            parts = line.strip().split(':')
            if len(parts) == 2:
                model_name = parts[0].strip()
                accuracy = float(parts[1].strip())
                accuracies[model_name] = accuracy
            else:
                print(f"Invalid format in line: {line}")
    return accuracies

# File path containing accuracies
file_path = "/content/drive/MyDrive/Accuracies.txt"

# Read accuracies from the file
accuracies = read_accuracies(file_path)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(accuracies.keys(), accuracies.values(), color='skyblue')
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.title("Test Accuracies of Different Models")
plt.ylim(0, 1)  # Set y-axis limit from 0 to 1 for better visualization
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
